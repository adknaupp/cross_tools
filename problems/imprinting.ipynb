# Questions on Imprinting, context
In 1991, DeChiara et al. (Cell 64: 849-59) reported the parental imprinting of the mouse Insulin-like Growth Factor II gene (IGF2). The gene is imprinted such that only one allele is active (monoallelic expression), depending on parental origin—and this pattern of expression is maintained epigenetically in almost all tissues. To study the inheritance of imprinting patterns, the author developed a mutant mouse carrying a 250 bp deletion that spanned the first coding exon of the IGF2 gene. The mutant allele (Δigf2) was transmitted through the germline. Mice with an active Δigf2 allele are fertile, but “growth deficient” (see figure). To examine the genetic basis for the parental imprinting, the authors initiated a breeding program by mating heterozygous F1 progeny of both sexes to homozygous wild-type partners:
