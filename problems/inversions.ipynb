# Context for inversions questions (6, 7 and 8)
Suppose that all individuals in a wild population of Drosophila melanogaster are homozygous for an inversion in chromosome 3 that spans 8 cM. Also suppose that all individuals in a second population are homozygous or hemizygous for an inversion in the X chromosome that spans 14 cM. A female from the first population is mated with a male from the second to produce an F1 generation. Note that recombination (crossing over) does not occur in Drosophila males.

## Question 6
As pertaining to the inversions, the female F1 progeny are best described as:


```python
# The correct answer is Balanced and heterozygous for inversions on Chromosomes 3 and X. Both parents are homozygous for their respective inversion. Thus, a female progeny will inherit an inverted segment from each parent and a wild type (non-inverted segment) from each parent. What do you think the male would look like? 

# answer is Balanced and heterozygous for inversions on chromosomes 3 and X 
```

## Question 7
Taking into account recombination (crossing over) and the size of the inversions, what proportion of the gametes from the F1 female should be unbalanced (for either chromosome)? 

```python
# The proportion of unbalanced gametes equals the proportion of crossover-type gametes, which can be determined from the recombination frequencies within the inversions. The female progeny are heterozygous for two inversions. The inversion in chromosome 3 spans 8 cM, so the probability of a crossover within that region is 8%, and the probability of no crossover in that region is 92%. The inversion in the X chromosome spans 14 cM, so the probability of a crossover in that region is 14%, and the probability of no crossover in that region is 86%. The two chromosomes assort independently and any gamete that arises from a crossover in either inversion is unbalanced. Hence the probability of no crossovers in either inversion is 0.92 X 0.86 = 0.791. The proportion of unbalanced gametes can most easily be determined by subtracting the proportion of balanced gametes from unity: 1 - 0.791 = 0.209. 

# answer is 0.209
```

## Question 8
 If the males and females generated by the cross (F1 generation) are allowed to randomly intermate, what portion of the next generation (F2) females should be heterozygous for both inversions? 


```python
# Viable F1 male gametes that will produce female F2 progeny include 3 X and 3i X, whereas viable F1 female gametes include 3 X, 3 Xi, 3i X and 3i Xi. I have used 3i and Xi to indicate the inverted versions of chromosomes 3 and X, respectively. If you construct a Punnett square you will find that two out of eight combinations of male and female gametes yield an F2 progeny genotype that is heterozygous for both inversions. 

# answer is 1/4
```
