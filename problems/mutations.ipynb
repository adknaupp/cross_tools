## Question 1
Which of the following amino acid changes could result from a single transversion event? 

Gln→Asp

Met→Thr

Ile→Val

Ser→Tyr

Leu→Pro

```python
# transversions include 
# C -> T or A
# G -> T or A
# T -> G or C
# A -> G or C

# Gln→Asp CAA/CAG -> GAT/GAC  

# Met→Thr ATG -> AC*.         _T_ -> _C_ = single transition  <------

# Ile→Val AT(TCA) -> GT*     AT* -> GT* = single transition

# Ser→Tyr AGT/AGC -> TAT/TAC   AG_  -> TA_ = multiple changes

# Leu→Pro CT* -> CC*          CT_ -> CC_ = single transition

# which stand gets transcribed??
# 5' AATCCGAGA 3'
# 3' TTAGGCTCT 5'

# DNA matches the tRNA sequence (the 'anticodon') except thymine -> uracil
# AAT (template strand)
# TTA (complement strand)
# UUA (mRNA codon)
# AAU (tRNA anticodon)
```
