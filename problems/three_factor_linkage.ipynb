## Question 3: linkage, conformation, interference, predict progeny phenotypes
In Drosophila melanogaster, black body (b) is recessive to gray body (b+), purple eyes (pr) are recessive to red eyes (pr+), and vestigial wings (vg) are recessive to normal wings (vg+). These three genes are linked with pr located between b and vg at a genetic distance of 6 cM from b and 13 cM from vg. Previous experiments have shown that the interference associated with these three genes is 0.5. A fly with a black body, purple eyes and vestigial wings is crossed with a fly with a gray body, red eyes and normal wings. All of their F1 progeny is wild-type. The F1 females are mated with male flies that have black bodies, purple eyes and vestigial wings. Ten thousand progeny are produced from this testcross. How many of the testcross progeny are expected to have purple eyes but be wild-type for body color and wing type?

```python
# In the cross described in this problem all three genes are arranged in coupling on the female F1 chromosome. Because purple is in the middle then any flies with purple eyes that are wild-type for the other two traits must have arisen from a double crossover. The frequency of double crossovers is the product of the recombination frequencies of the two map intervals (0.06 x 0.13 = 0.0078). Out of 10,000 flies, 78 are expected to arise from double crossovers. Yet the stated interference is 0.5, so we only expect to see half of the calculated double crossovers or 39. Of these, half will have purple eyes, gray bodies and normal wings and the other half will have black bodies, red eyes and vestigial wings. Half of 39 is 19.5.  

# answer is 19.5

```

## Question 6: gene order
What is the correct order of these genes on the X chromosome? 

```python
# The two parental types must be white, bifid, which came from the female parent and rudimentary, which came from the male parent. The white, bifid phenotype is the most abundant in the F2 progeny but the rudimentary phenotype is much less abundant because of the reduced viability associated with the r allele. The double crossover can be difficult to identify because three phenotypes have just 1 or 2 flies. To get around this problem, look for the least abundant phenotype that does not include rudimentary. This would be white, which only has 2 flies. This is a double crossover phenotype. The reciprocal phenotype is bifid, rudimentary and so it is a double crossover phenotype also. Because bi is paired with r in the double crossover genotype, then bi must be the gene in the middle and the correct order is r bi w. 

# answer is r bi  w 
```

## Question 7: recombination frequency
What is the observed recombination frequency between bi and w?

```python
# The phenotypes resulting from a single crossover between w and bi are bifid and white, rudimentary of which there are 10 and 2 flies, respectively. The double crossover types are white and bifid, rudimentary of which there are 2 and 1 flies, respectively. The total number of recombinant phenotypes resulting from a crossover between bi and w are 10 + 2 + 2 + 1 = 15. Dividing this by 395, which is the total number of flies scored, gives a recombination frequency of 0.038. 

# answer is 0.038
```

## Question 8: recombination frequency, interference
What is the interference calculated from these data? Use the uncorrected recombination frequency and round your answer to two decimal places.

```python
# Interference is defined as 1 - C where C is the coefficient of coincidence or the number of observed double crossovers divided by the number of expected double crossovers. To calculate the number of expected double crossovers, multiply the frequency of each of the two single crossovers and then multiply this product by the total number of flies that were scored.  For this problem, the recombination frequency between bi and w is 0.038 and between bi and r it is 0.423.  The coefficient of coinicidence (C) is 0.038 x 0.423 x 395 = 0.47.  Interference is 1 - C = 1 - 0.47 = 0.53.

# answer is 0.53
```

## Question 10: genetic distance, kosambi
Among the experiments listed in Sturtevant’s paper was a cross between a white-eyed, miniature-winged female and a wild-type male. All of the F1 females were wild type, whereas all of the F1 males had white eyes and miniature wings. In the F2 generation there were 395 wild-type flies, 223 flies with white eyes, 247 with miniature wings and 382 with white eyes and miniature wings. 

What is the genetic map distance, corrected using the Kosambi function, between white and miniature? 

```python
# The white-eyed flies and the miniature-winged flies are the recombinants, of which there are 223 and 247, respectively. The total number of recombinants is 470. The total number of F2 flies scored are 1247; hence, the recombination frequency is 470/1247, which is equal to 0.377. To calculate genetic map distance use the Kosambi function to correct for double crossovers and interference. The function is x = 0.25ln[(1 + 2y)/(1 – 2y)], where y is the observed recombination frequency of 0.377 and x is the corrected frequency. For this problem, x = 0.491. Multiply x by 100 to arrive at the corrected map distance. 

# answer is 49.1 cM
```
