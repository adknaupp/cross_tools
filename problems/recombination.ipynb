## Pedigree data featuring one recombinant phenotype
Nail-patella syndrome is an autosomal dominant disorder in humans that results in abnormal nail growth and absent or severely reduced patella. The mutation occurs in the NPS1 gene located on chromosome 9. In 1955, Renwick and Lawler (Ann. Hum. Genet. 19: 312-331) were the first to demonstrate linkage between two human genes using pedigree analysis when they reported linkage between NPS1 and the ABO gene, which control the ABO blood type. The pedigree to the right shows three generations of a family with members who have Nail-patella syndrome. Shaded symbols show individuals with the syndrome and the blood type of each person in the family is given inside of their respective symbol.
